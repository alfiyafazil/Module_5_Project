```python
import pandas as pd

filename = "automobile.data"
df = pd.read_csv(filename)
print("The first 10 rows of the dataframe") 
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
df.head(10)
```

```python
import numpy as np

df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
df[['price']].head(11)
```

```python
df.dropna(subset=["price"],axis=0,inplace=True)
```

```python
df[['price']].head(10)
```

```python
df[['bore']].loc[53:61]
```

```python
mean = df[['bore']].astype("float").mean(axis=0)
mean
```

```python
df[['bore']] = df[['bore']].replace(np.nan,mean)
```

```python
df[['bore']].loc[53:61]
```

```python
df[['price']].dtypes
```

```python
df[["price"]] = df[["price"]].astype("float")
```

```python
df[['price']].dtypes
```

```python
df["city-mpg"] = 235 / df["city-mpg"]
df.rename(columns={'city-mpg':'city-L/100km'}, inplace=True)
df["highway-mpg"] = 235/df["highway-mpg"]
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
df.head()
```

```python
df["length"] = df["length"]/df["length"].max()
df["width"] = df["width"]/df["width"].max()
df['height'] = df['height']/df['height'].max() 
df[["length","width","height"]].head()
```

```python
binwidth = int((max(df['price'])-min(df['price']))/3)
```

```python
bins = range(int(min(df['price'])),int(max(df['price'])),binwidth)
```

```python
df['price-binned'] = pd.cut(df['price'],bins, labels=["Low","Medium","High"])
```

```python
df.loc[15:20,['price','price-binned']]
```

```python
df['price-binned'].dtypes
```

```python
df.to_csv('clean_data.csv')
```
